** Step 3: Generate Operational Insights**

Now that the data is clean, we can start using SQL to extract meaningful business insights.

* Find out which payment methods are most frequently used by customers
* Identify the top 10 cities with the highest number of orders

 Task:
In this step, we will:

Use SQL queries to analyze the cleaned dataset

In this step, we will use SQL queries to generate operational insights.

First, group the data by payment mode to count how many times each mode was used

Then, group the data by city to count how many orders came from each city

Finally, we will use Python slicing to show only the top 10 cities by order volume

This will help the business understand popular payment methods and high-performing


```python
# Install pandasql if not already installed
!pip install pandasql

# Import the necessary library
from pandasql import sqldf

# Define a function to execute SQL queries on the DataFrame
# This function is needed by pandasql
pysqldf = lambda q: sqldf(q, globals())

# --- Operational Insight 1: Most frequently used payment methods ---

# SQL query to count the usage of each payment method
# Assuming the column name for payment method is 'PaymentMode' based on the user's description
q_payment_methods = """
SELECT
    "PaymentMode",
    COUNT(*) AS OrderCount
FROM
    df
GROUP BY
    "PaymentMode"
ORDER BY
    OrderCount DESC;
"""

# Execute the query
payment_methods_df = pysqldf(q_payment_methods)

# Display the results
print("\nMost Frequently Used Payment Methods:")
display(payment_methods_df)

# --- Operational Insight 2: Top 10 cities with the highest number of orders ---

# SQL query to count orders per city and get the top 10
# Assuming the column name for city is 'City' based on the user's description
q_top_cities = """
SELECT
    "City",
    COUNT(*) AS OrderCount
FROM
    df
GROUP BY
    "City"
ORDER BY
    OrderCount DESC
LIMIT 10;
"""

# Execute the query
top_cities_df = pysqldf(q_top_cities)

# Display the results
print("\nTop 10 Cities by Order Volume:")
display(top_cities_df)
```

** Step 4: Product Analysis**

* Products that are expensive but sold in small quantities (maybe overpriced or not promoted well)
* Categories that are most popular (so they know whats in demand)

 Task:
In this step, we will use SQL to analyze product performance. Youll help the company identify:

Group the data by sub-category, calculate average price and total quantity sold, and sort the results to find premium-priced products with low sales

Next, group the data by product category and sort by total quantity sold to identify most popular product categories

These insights help identify products that may need pricing adjustments or promotion, and categories that are high in demand.

```python
# Import the necessary library (already imported in the previous step, but good practice to include)
from pandasql import sqldf

# Define a function to execute SQL queries on the DataFrame (already defined, but good practice to include)
pysqldf = lambda q: sqldf(q, globals())

# --- Product Analysis 1: Expensive products with low sales ---

# SQL query to analyze products by sub-category
# Calculate average 'Amount' (assuming 'Amount' represents price or total sale for the item)
# Calculate total 'Quantity' sold
# Assuming the column names are 'Sub-Category', 'Amount', and 'Quantity'
q_premium_low_sales = """
SELECT
    "Sub-Category",
    AVG("Amount") AS AveragePrice, -- Assuming 'Amount' is the price for analysis
    SUM("Quantity") AS TotalQuantitySold
FROM
    df
GROUP BY
    "Sub-Category"
ORDER BY
    AveragePrice DESC, -- Sort by average price descending
    TotalQuantitySold ASC; -- Then sort by total quantity sold ascending to find low sales
"""

# Execute the query
premium_low_sales_df = pysqldf(q_premium_low_sales)

# Display the results
print("\nSub-Categories by Average Price and Total Quantity Sold (Sorted by Price DESC, Quantity ASC):")
display(premium_low_sales_df)


# --- Product Analysis 2: Most popular product categories ---

# SQL query to identify most popular product categories
# Group by 'Category' and sum the 'Quantity' sold
# Assuming the column names are 'Category' and 'Quantity'
q_popular_categories = """
SELECT
    "Category",
    SUM("Quantity") AS TotalQuantitySold
FROM
    df
GROUP BY
    "Category"
ORDER BY
    TotalQuantitySold DESC; -- Sort by total quantity sold in descending order
"""

# Execute the query
popular_categories_df = pysqldf(q_popular_categories)

# Display the results
print("\nMost Popular Product Categories (Sorted by Total Quantity Sold DESC):")
display(popular_categories_df)
```

** Step 5: Data Visualization**

** Plot 1: Top 10 Products by Revenue**

 Task:
We want to find which products generate the most money.

To do this:

Use groupby() on the Sub-Category column

Sum the Amount column to get total revenue per product type

Sort the results in descending order

Show the top 10 products using a vertical bar chart

This will help identify which products drive the most revenue.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Data Visualization 1: Top 10 Products by Revenue ---

# Group data by 'Sub-Category' and sum 'Amount' to get total revenue per sub-category
# Assuming 'Amount' represents the revenue for each transaction line item
revenue_by_subcategory = df.groupby('Sub-Category')['Amount'].sum().reset_index()

# Sort the results in descending order based on total revenue
top_10_subcategories = revenue_by_subcategory.sort_values(by='Amount', ascending=False).head(10)

# Display the top 10 sub-categories by revenue
print("Top 10 Sub-Categories by Revenue:")
display(top_10_subcategories)

# Create a vertical bar chart for the top 10 sub-categories by revenue
plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
sns.barplot(x='Amount', y='Sub-Category', data=top_10_subcategories, palette='viridis')

plt.title('Top 10 Sub-Categories by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Sub-Category')
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

** Plot 2: Monthly Order Trend**

 Task:
We want to see how the number of orders changes month by month.

To do this:

Use the Year-Month column

Count how many orders happened each month using value_counts()

Sort the months so they appear in the correct order

Plot this as a line chart to visualize changes over time

This helps the business understand seasonal trends, growth, or any decline in performance.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame and it has a 'Year-Month' column
# (If your DataFrame is named differently or the column is named differently, adjust the code)

# 1. Count how many orders happened each month
monthly_orders = df['Year-Month'].value_counts()

# 2. Sort the months
monthly_orders_sorted = monthly_orders.sort_index()

# 3. Plot this as a line chart
plt.figure(figsize=(12, 6)) # Adjust figure size as needed
monthly_orders_sorted.plot(kind='line')

plt.title('Monthly Order Trend')
plt.xlabel('Year-Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.grid(True) # Add a grid for better readability

# Display the plot
plt.show()
```

** Plot 3: Top 10 States by Total Profit**

 Task:
We want to find which states are generating the most profit for the company.

To do this:

Group the data by State

Sum the Profit for each state

Sort the results in descending order

Plot the top 10 states using a horizontal bar chart

This helps the business decide which states to invest more in or review if low-performing.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame and it has 'State' and 'Profit' columns
# (If your DataFrame is named differently or the columns are named differently, adjust the code)

# 1. Group the data by State and sum the Profit
state_profit = df.groupby('State')['Profit'].sum()

# 2. Sort the results in descending order
top_10_states = state_profit.sort_values(ascending=False).head(10)

# 3. Plot the top 10 states using a horizontal bar chart
plt.figure(figsize=(10, 7)) # Adjust figure size as needed
top_10_states.plot(kind='barh')

plt.title('Top 10 States by Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('State')
plt.gca().invert_yaxis() # Invert y-axis to have the highest profit at the top
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Display the plot
plt.show()
```

To do this:

Group the dataset by PaymentMode

Calculate the total profit for each payment type

Sort the results in descending order

Plot the totals using a Seaborn bar plot

** Step 7: Export Business Reports**

 Task:
In this final step, you will create and export a summary report that combines two different business insights:

** 1. Top 10 Products by Revenue**
* Group the data by both Category and Sub-Category.
* Calculate the total revenue using the Amount column.
* Sort the results in descending order of revenue.
* Select the top 10 entries using .head(10).
* Reset the index to make the grouped values into columns.
* Rename the columns to clearly reflect:
["Top Product Category", "Sub-Category", "Total Revenue"].

** 2. Monthly Order Volume**
* Use the Year-Month column to count the number of orders per month.
* Sort the months in chronological order.
* Convert the result into a DataFrame using .reset_index().
* Rename the columns as:
["Month", "Number of Orders"].

Take only the first 10 months to match the number of rows from the Top Products table.

** 3. Combine and Export**
* Use pd.concat(..., axis=1) to join both tables side by side.
* Export the final combined DataFrame to a CSV file named: "top_products_and_monthly_orders.csv" using .to_csv().

```python
import pandas as pd

# Assuming 'df' is your DataFrame with 'Category', 'Sub-Category', 'Amount', and 'Year-Month' columns
# (Adjust DataFrame and column names as needed)

# 1. Top 10 Products by Revenue
top_products_revenue = df.groupby(['Category', 'Sub-Category'])['Amount'].sum() \
    .sort_values(ascending=False) \
    .head(10) \
    .reset_index()

top_products_revenue.columns = ["Top Product Category", "Sub-Category", "Total Revenue"]

# 2. Monthly Order Volume
monthly_order_volume = df['Year-Month'].value_counts() \
    .sort_index() \
    .reset_index()

monthly_order_volume.columns = ["Month", "Number of Orders"]

# Take only the first 10 months to match the number of rows from the Top Products table
monthly_order_volume_top10 = monthly_order_volume.head(10)

# 3. Combine and Export
combined_report = pd.concat([top_products_revenue, monthly_order_volume_top10], axis=1)

# Export the final combined DataFrame to a CSV file
combined_report.to_csv("top_products_and_monthly_orders.csv", index=False)

print("Business report 'top_products_and_monthly_orders.csv' created successfully.")
```
