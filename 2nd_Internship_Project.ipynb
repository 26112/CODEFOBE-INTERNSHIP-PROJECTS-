## **Step by Step Problem**
**Step 1: Load and Explore the Dataset**

Before analyzing the sales data, we need to load the dataset into Pandas and explore its structure.

 Task: Load the CSV File and Display Basic Information

```python
# solution code for step 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the CSV file
df = pd.read_csv('Retail_sales_data.csv')

# Display the first few rows
print(df.head())
# Display dataset structure
print(df.info())
```

**Step 2: Data Cleaning & Preprocessing**

Handle missing values, correct data types, and ensure consistency:

* Replace missing values in numerical columns (Quantity Sold, Unit Price, Total Sales Amount) with mean
* Replace missing values in categorical columns (Payment Method, Customer Segment) with mode
* Convert Date column to datetime
* Check and confirm correct data types using .dtypes

This step ensures the dataset is clean, consistent, and ready for analysis.

```python
# Solution code for step 2
# List of numerical columns
numerical_cols = ['Quantity Sold', 'Unit Price', 'Total Sales Amount']

# Replace missing values with mean
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)
# List of categorical columns
categorical_cols = ['Payment Method', 'Customer Segment']

# Replace missing values with mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
    # Convert Date column to datetime
# Explicitly specify the date format as 'day-month-year'
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
# Check data types
print(df.dtypes)
```

** Step 3: Operational Insight**

Identify key business trends that help management optimize operations:
* Most common payment methods used by customers
* Cities with the highest number of orders


 1. Identify the Most Common Payment Methods used by customers

```python
# Solution code for step 3: part 1
# Most common payment methods
payment_method_counts = df['Payment Method'].value_counts()
print(payment_method_counts)
```

 2. Identify Cities with the Highest Order Counts

```python
# Solution code for Step 3 part 2
# Cities with the highest number of orders
city_order_counts = df['City'].value_counts()
print(city_order_counts)
```

** Step 4 Product Analysis**

Detect trends and inefficiencies in product sales and pricing:

Products with premium pricing but low sales volume

Categories most frequently purchased by different customer segments

 1. Products with Premium Pricing but Low Sales Volume

 Goal:
Find products that have a high unit price but low total quantity sold  possibly overpriced or not in demand.


```python
# Solution code for Step 4 part 1
# Calculate 75th percentile of unit price
premium_price_threshold = df['Unit Price'].quantile(0.75)

# Calculate 25th percentile of quantity sold
low_sales_threshold = df['Quantity Sold'].quantile(0.25)

# Find products with premium pricing and low sales
premium_low_sales_products = df[(df['Unit Price'] > premium_price_threshold) & (df['Quantity Sold'] < low_sales_threshold)]
print(premium_low_sales_products[['Product Name', 'Unit Price', 'Quantity Sold']])
```

 2. Categories Most Frequently Purchased by Different Customer Segments

 Goal:
See which product categories are most popular for each customer segment (e.g., Corporate, Individual).




```python
# Solution code for Step 4 part 2
# print(df.columns) # Keep this line to verify the column names if needed
category_preference = df.groupby(['Customer Segment', 'Category']).size().unstack(fill_value=0)
print(category_preference)
```

**Step 5: Customer Behavior**

These insights help understand who your customers are, how much they spend, and how often they shop.

Use the dataset to reveal customer behavior patterns:
* Identify top-spending customers by total sales
* Track repeat buyers vs one-time buyers

 1. Identify Top-Spending Customers by Total Sales

 Goal:
Find customers who have spent the most overall (high-value customers).

```python
# Solution code for Step 5 part 1
# Group by customer and sum total sales
top_spending_customers = df.groupby('Customer ID')['Total Sales Amount'].sum().sort_values(ascending=False)
print(top_spending_customers.head(10))  # Top 10 customers
```

 2. Track Repeat Buyers vs One-Time Buyers

 Goal:
Determine how many customers placed multiple orders vs only one order.

```python
# Solution code for Step 5 part 2
# Group by customer and count the number of orders
customer_order_counts = df.groupby('Customer ID').size()
# one time buyers - customers who have bought for a single time (orderid == 1)
# repeat buyers - customers who have bought for multiple times (orderid > 1).
# One-time buyers
one_time_buyers = customer_order_counts[customer_order_counts == 1].count()

# Repeat buyers
repeat_buyers = customer_order_counts[customer_order_counts > 1].count()

print(f"One-time buyers: {one_time_buyers}")
print(f"Repeat buyers: {repeat_buyers}")
```

** Step 6 Data Visualization**

Present your insights using clean and business-relevant visuals:
 1. Growth Line Plot  Daily Order Volume Over Time
 2. Vertical Bar Plot  Top 10 Products by Total Revenue
 3. Count Plot  Orders by Customer Segment

 1. Growth Line Plot  Daily Order Volume Over Time

```python
# Solution code for Step 6 part 1'
# Aggregate daily order volume
daily_orders = df.groupby(df['Date'].dt.date).size()

# Plot
plt.figure(figsize=(12,6))
daily_orders.plot(kind='line')
plt.title('Daily Order Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

 2. Vertical Bar Plot  Top 10 Products by Total Revenue

```python
# Solution code for Step 6 part 2
# Calculate total revenue per product
product_revenue = df.groupby('Product Name')['Total Sales Amount'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
product_revenue.plot(kind='bar')
plt.title('Top 10 Products by Total Revenue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.show()
```

 3. Count Plot  Orders by Customer Segment



```python
# Solution code for Step 6 part 3
# Count plot for orders by customer segment
plt.figure(figsize=(10,6))
sns.countplot(x='Customer Segment', data=df)
plt.title('Orders by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Orders')
plt.show()
```

**Step 7: Export Business Reports (CSV)**

Save your results for managers and decision-makers to access:

* Export Country-wise Average Order Value (AOV)

All files should be saved in .csv format.

 1. Export Top 10 Products by Revenue

```python
# Solution code for Step 7 part 1
# Export top 10 products by revenue
product_revenue.to_csv('top_10_products_by_revenue.csv')
```

 2. Export Country-wise Average Order Value (AOV)

```python
# Solution code for Step 7 part 2
# Calculate AOV per country
aov_per_country = df.groupby('Country')['Total Sales Amount'].mean()

# Export to CSV
aov_per_country.to_csv('country_wise_aov.csv')
```
